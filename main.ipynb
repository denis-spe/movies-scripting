```r
home <- read_html("http://movies2watch.tv/movie")
movies <- home %>% rvest::html_elements(css = ".film_list-wrap a")
link <- movies %>% html_attr(name="href")
mapped_link <- link %>% map(function(x) sprintf("%s%s", "http://movies2watch.tv", x)) %>% unlist()
```

```r
movie.detail <- function(movie_link) {
movie.page <- read_html(movie_link)
detail <- movie.page %>% html_elements(css = ".row-line")
title <- movie.page %>% html_element(css = ".heading-name") %>% html_text2()
movie.info <- detail %>% html_text2()

info <- movie.info %>% map(function(x){
    lst <- list() 
    strs <- x %>% str_split(": ") %>% unlist()

    lst[strs[[1]]] <- strs[[2]]

    lst
})

df <- info %>% as.data.frame()
df[["Title"]] <- title

df

}
```

```r
movie.pages <- function(page_num, csv_name="movies.csv"){
    url <- sprintf("http://movies2watch.tv/movie?page=%d", page_num)
    home <- read_html(url)
    movies <- home %>% rvest::html_elements(css = ".film_list-wrap a")
    link <- movies %>% html_attr(name="href")
    mapped_link <- link %>% map(function(x) sprintf("%s%s", "http://movies2watch.tv", x)) %>% unlist()
    dfs <- mapped_link %>% map(movie.detail)
    df <- do.call(rbind, dfs) %>% distinct()

    if (file.exists(csv_name)){
        movie_df <- read.csv(csv_name)

        recreated_df <- rbind(df, movie_df) %>% distinct()
        write.csv(recreated_df, csv_name, row.names=F)
        name <- str_split(csv_name, pattern = "[\\.]")[[1]][1]
        print(sprintf("Updated the %s with the %d page", name, page_num))
        
    } else {
       write.csv(df, csv_name, row.names=F)
       print(sprintf("successfully created %s", csv_name))
    }

}
```

```r
for (idx in 73:500) movie.pages(idx)
```
